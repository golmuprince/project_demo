```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Optional, Literal
```

```python
class AgentState(TypedDict):
    mode: str
    user_input: str 
    exam_date: Optional[str]
    study_plan: Optional[dict]
    current_topic: Optional[str]
    quiz_question: Optional[str]
    quiz_answer: Optional[str]
    message: str

graph_builder = StateGraph(AgentState)
```

```python
def router_node(state: AgentState) -> AgentState:
    return state

def planner_node(state: AgentState) -> AgentState:
    return state

def class_node(state: AgentState) -> AgentState:
    return state

def quiz_node(state: AgentState) -> AgentState:
    return state

def route_decision(state: AgentState) -> Literal["planner", "class", "quiz"]:
    mode = state.get("mode", "planner")
    if mode in ["planner", "class", "quiz"]:
        return mode
    return "planner" 

graph_builder.add_node("router", router_node)
graph_builder.add_node("planner", planner_node)
graph_builder.add_node("class", class_node)
graph_builder.add_node("quiz", quiz_node)

graph_builder.add_edge(START, "router")
graph_builder.add_conditional_edges(
    "router",
    route_decision, 
    {
        "planner": "planner",
        "class": "class",
        "quiz": "quiz",
    }
)
graph_builder.add_edge("planner", END)
graph_builder.add_edge("class", END)
graph_builder.add_edge("quiz", END)

graph = graph_builder.compile() 
```

```python
print(graph.get_graph().draw_ascii())
```
